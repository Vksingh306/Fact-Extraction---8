```python
import pandas as pd
import numpy as np
from numpy import inf,NaN
import json
import pickle
```

```python
def dumpData(d, f):
    fd = open(f, 'ab')
    pickle.dump(d, fd)
    fd.close()
def loadData(pickle_file):
    fd = open(pickle_file, 'rb')
    d = pickle.load(fd)
    fd.close()
    return d
```

```python
df = pd.read_csv('weight-8-sections.csv' , index_col = 0)
```

```python
section_case_dict = loadData('section_case_dict')
```

```python
dict_chosen = {}
```

```python
for i in df.index:
    lst_tmp = [z for z in df.index if z != i]
    lst_tmp2 = []
    for j in lst_tmp:
        for z in section_case_dict[j]:
            lst_tmp2.append(z)
    set_tmp = set(lst_tmp2)
    lst_minus = list(set_tmp)
    dict_chosen[i] = [x for x in section_case_dict[i] if x not in lst_minus]
```

```python
for k,v in dict_chosen.items():
    print(k,len(v))
```
