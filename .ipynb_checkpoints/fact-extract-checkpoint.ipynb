```python
def dumpData(d, f):
    fd = open(f, 'ab')
    pickle.dump(d, fd)
    fd.close()
def loadData(pickle_file):
    fd = open(pickle_file, 'rb')
    d = pickle.load(fd)
    fd.close()
    return d
```
