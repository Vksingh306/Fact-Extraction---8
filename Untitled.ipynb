```python
for (sec_1,sec_2) in combinations(sections_cases.keys(),2):
    sections_common[f"{sec_1}-{sec_2}"] = set(sections_cases[sec_1]).intersection(set(sections_cases[sec_2]))
```

```python
mean = []
stdev = []
for i in nx.find_cliques(G_2):
    print(i)
    edge_weight_tmp = []
    for sec_1,sec_2 in combinations(i,2):
        try:
            edge_weight_tmp.append(len(sections_common[f"{sec_1}-{sec_2}"]))
        except:
            edge_weight_tmp.append(len(sections_common[f"{sec_2}-{sec_1}"]))
    mean.append(stat.mean(edge_weight_tmp))
    stdev.append(stat.pstdev(edge_weight_tmp))

print(mean , stdev)
```
