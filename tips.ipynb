# Indexer and Retriever

```python
# import csv
import pandas as pd
import nltk
from nltk import word_tokenize
porter = nltk.PorterStemmer()
```

```python
file_path = '/home/dwaipayan/Dropbox/courses/2021-irws/datasets/wiki_movie_plots_deduped.csv'
```

```python
def make_inverted_index(file_path):
    df = pd.read_csv(file_path)
    df = df['Plot']
    inverted_index = {}
    # this is a dictionary with key - term and value - [df(term), pl(term)]

    docid = 0
    for movie in df:
        print("Movie -", docid)
        tokens = word_tokenize(movie)
        tokens_lower = [w.lower() for w in tokens if w.isalnum()]
        stemmed = [porter.stem(w) for w in tokens_lower]
        
        for stem in stemmed:
            val = inverted_index.get(stem)
            if val == None:
                l = [1, [docid]]
                inverted_index[stem] = l
            else:
                l = inverted_index[stem]
                # l[0] = df(stem)
                # l[1] = pl(stem)
                if docid not in l[1]:
                    l[1].append(docid)
                    l[0] += 1
        docid += 1
    return inverted_index
```

```python
inverted = make_inverted_index(file_path)
```

### Pickle

```python
import pickle

def dumpData(d, f):
    fd = open(f, 'ab')
    pickle.dump(d, fd)
    fd.close()
def loadData(pickle_file):
    fd = open(pickle_file, 'rb')
    d = pickle.load(fd)
    fd.close()
    return d
```

```python
dumpData(inverted, 'inverted_dump.pickle')
```

# Retriever

```python
def processText(text):
    tokens = word_tokenize(text)
    tokens_lower = [w.lower() for w in tokens if w.isalnum()]
    stemmed = [porter.stem(w) for w in tokens_lower]
    return stemmed
```

```python
Q = input("Enter the query: ")
Q = Q.lower()
Q = processText(Q)
```

```python
print(Q)
```

```python
all_pl = []
for q in Q:
    pl = inverted.get(q)
    if pl is not None:
        all_pl.append(pl[1])
```

```python
print(all_pl)
```

```python
print(set.intersection(*map(set, all_pl)))
```

```python
print(set.union(*map(set, all_pl)))
```
